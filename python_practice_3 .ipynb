Але якщо число непарне, помножте його на 3 та додайте 1 і виведіть на екран.

5. Створіть словник, використовуючи dict comprehension, що містить if else. Коли слід використовувати comprehensions, а коли їх слід уникати?

```python
# Створення словника, де ключі - це числа від 0 до 9, а значення - 'even' або 'odd' в залежності від парності
my_dict = {x: 'even' if x % 2 == 0 else 'odd' for x in range(10)}

# Виведення результату
print(my_dict)
```

Коли використовувати comprehensions: Якщо  проста умова, яка може бути виражена одним рядком, створювати списки, словники та множини для простих операцій.Коли уникати : Якщо  умова стає складною та важко читається в одному рядку, вираження має багато вкладень, це може призвести до погіршення читабельності.

6. Створіть список з рядками та запишіть його у змінну. Створіть ще один список на основі попереднього, де кожен рядок буде містити тільки три перших заглавних символи. Наприклад: ['hello', 'i', 'dont', 'care'] -> ['HEL', 'I', 'DON', 'CAR']. Використання comprehensions обовʼязкове.

```python
# Створення початкового списку з рядками
original_list = ['hello', 'i', 'dont', 'care']

# Використання list comprehension для створення нового списку з триперших заголовних символів кожного рядка
modified_list = [word[:3].upper() for word in original_list]

# Виведення результату
print(modified_list)
```

7. Продемонструйте роботу циклу while. Не можна використовувати True для запуску циклу, щоб уникнути безкінечного циклу!

```python
# Початкове значення лічильника
counter = 1

# Цикл while, який виконується, доки лічильник менший або рівний 5
while counter <= 5:
    print(f"Iteration {counter}")
    counter += 1  # Збільшення лічильника на 1 після кожної ітерації

# Виведення повідомлення після завершення циклу
print("Цикл завершено.")
```

8. Напишіть програму, яка виведе на екран непарні числа в діапазоні від 0 до 20 включно.

9. Створіть список з елементами булевого типу або None. Використовуючи цикли, отримайте в результаті список з кортежами, де перший елемент кортежу - індекс(ціле число), а другий елемент - відповідне значення з першого списку.

Наприклад, [True, True, None, False] -> [(0, True), (1, True), (2, None), (3, False)]

```python
# Початковий список з елементами булевого типу або None
original_list = [True, True, None, False]

# Створення списку з кортежами за допомогою циклу
result_list = []
for index, value in enumerate(original_list):
    result_list.append((index, value))

# Виведення результату
print(result_list)
```

10. Створіть словник, де ключі - назви книжок, а значення - їхня кількість у наявності в Вашій міні-бібліотеці. Бібліотека має містити щонайменше 6 книжок (6 пар значень у словнику) і щонайбільше 10 (пар значень). Використовуючи цикли, оновіть словник (не створюйте новий) так, щоби кількість книг у наявності збільшилося на 5 кожної книги. Наприклад,
{'It': 3, 'Fault stars': 10, 'Bible': 17, 'Psychological romance': 4, 'Harry Potter': 13} -> {'It': 8, 'Fault stars': 15, 'Bible': 22, 'Psychological romance': 9, 'Harry Potter': 18}

```python
# Початковий словник з кількістю книг у наявності
library = {
    'It': 3,
    'Fault stars': 10,
    'Bible': 17,
    'Psychological romance': 4,
    'Harry Potter': 13,
    'Python Programming': 8,
}

# Оновлення словника, збільшуючи кількість книг у наявності на 5
for book in library:
    library[book] += 5

# Виведення оновленого словника
print(library)
```

12. Гра "Нумо вгадай".
Розробіть просту гру, де користувач має вгадати випадкове число від 1 до 100. Для генерації випадкового числа використовуйте функцію random.randint(a, b) (тут a та b включно). Для отримання числа з консолі використовуйте функцію input(), результат якої обовʼязково явно приведіть до типу int.

Підказка: використовуйте цикл while та умови if-else, щоб повідомити користувачу піказки (напр. "Більше", "Менше") поки користувач не вгадає. По завершенню виведіть кількість спроб, які знадобилися для вгадування.

```python
import random

# Генерація випадкового числа від 1 до 100
secret_number = random.randint(1, 100)

# Лічильник спроб
attempts = 0

print("Вгадайте число від 1 до 100")

# Цикл для вгадування числа
while True:
    # Введення числа користувачем та перевірка на правильність введення
    guess = input("Ваша спроба: ")

    try:
        guess = int(guess)
    except ValueError:
        print("Будь ласка, введіть ціле число.")
        continue

    # Збільшення лічильника спроб
    attempts += 1

    # Перевірка, чи вгадав користувач
    if guess == secret_number:
        print(f"Вітаємо! Ви вгадали число {secret_number} за {attempts} спроб.")
        break
    elif guess < secret_number:
        print("Більше")
    else:
        print("Менше")
```
