```python
class Figure:
    def get_area(self):
        return 0

    def __add__(self, other):
        if isinstance(other, Figure):
            return self.get_area() + other.get_area()
        else:
            raise ValueError("Unsupported operand type")
```

6. (11б). Створіть 2 дочірніх класи від Фігури: Трикутник та Коло, які мають атрибути, необхідні для розрахунку площин. Визначте метод get_area в кожному з них так, щоби вони розраховували площу в залежності від формули для кожного типу фігури. Створіть обʼєкт класу Трикутник та обʼєкт класу Коло. Виконайте операцію суми за допомогою оператора + між ними.

```python
class Figure:
    def get_area(self):
        return 0

    def __add__(self, other):
        if isinstance(other, Figure):
            return self.get_area() + other.get_area()
        else:
            raise ValueError("Unsupported operand type")

class Triangle(Figure):
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def get_area(self):
        return 0.5 * self.base * self.height

class Circle(Figure):
    def __init__(self, radius):
        self.radius = radius

    def get_area(self):
        return 3.14 * self.radius**2

# Створення об'єктів класів Triangle та Circle
triangle = Triangle(base=6, height=4)
circle = Circle(radius=5)

# Виклик методу get_area для отримання площі кожної фігури
area_triangle = triangle.get_area()
area_circle = circle.get_area()

# Виклик методу __add__ для отримання суми площин фігур
sum_area = triangle + circle

# Виведення результатів
print(f"Площа трикутника: {area_triangle}")
print(f"Площа кола: {area_circle}")
print(f"Сума площ фігур: {sum_area}")
```

8. (13б). Створіть клас BankAccount з приватними атрибутами balance та account_number. Реалізуйте методи поповнення та зняття коштів, забезпечивши належну інкапсуляцію. Підказка: використовуйте декоратори property та setter.

```python
class BankAccount:
    def __init__(self, initial_balance, account_number):
        self._balance = initial_balance
        self._account_number = account_number

    @property
    def balance(self):
        return self._balance

    @property
    def account_number(self):
        return self._account_number

    @balance.setter
    def balance(self, new_balance):
        if new_balance >= 0:
            self._balance = new_balance
        else:
            print("Неприпустиме значення для балансу. ")

    def deposit(self, amount):
        if amount > 0:
            self._balance += amount
            print(f"Внесено {amount} грн. Новий баланс: {self._balance} грн.")
        else:
            print("Сума для внесення повинна бути більше 0.")

    def withdraw(self, amount):
        if amount > 0 and amount <= self._balance:
            self._balance -= amount
            print(f"Знято {amount} грн. Новий баланс: {self._balance} грн.")
        else:
            print(" недостатньо коштів на рахунку.")
```

10. (13б). Створіть клас Person з атрибутами name та age. Створіть ще один клас Employee з такими атрибутами, як department та salary. Створіть клас Manager, який успадковує обидва класи Person та Employee. Продемонструйте використання множинної спадковості, створивши обʼєкт класу Manager та отримавши mro для цього класу.

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display_info(self):
        print(f"Person: {self.name}, {self.age} years old")

class Employee:
    def __init__(self, department, salary):
        self.department = department
        self.salary = salary

    def display_info(self):
        print(f"Employee: {self.department} department, Salary: {self.salary}")

class Manager(Person, Employee):
    def __init__(self, name, age, department, salary):
        Person.__init__(self, name, age)
        Employee.__init__(self, department, salary)

    def display_info(self):
        print(f"Manager: {self.name}, {self.age} years old, {self.department} department, Salary: {self.salary}")

manager = Manager(name="Слєпченко", age=20, department="НаУКМА", salary=1000)

manager.display_info()

print(f"MRO for Manager: {Manager.mro()}")
```

```python
# Визначення рядка
string = "qwer"

# Створення ітератора
iterator = iter(string)

# Виклик функції next 5 разів
for _ in range(5):
    try:
        next_value = next(iterator)
        if _ < 4:
            print(next_value)
    except StopIteration:
        print("Досягнуто кінця послідовності")
        break
```

13. (11б). Створіть клас, що має визначений Iterator Protocol та при кожному виклику next повертає літери англійської абетки, поки вони не вичерпаються.

```python
class AlphabetIterator:
    def __init__(self):
        self.current_char = ord('a')

    def __iter__(self):
        return self

    def __next__(self):
        if self.current_char > ord('z'):
            raise StopIteration
        else:
            char = chr(self.current_char)
            self.current_char += 1
            return char

# Створення об'єкту ітератора
alphabet_iterator = AlphabetIterator()

# Використання ітератора
for _ in range(26):
    print(next(alphabet_iterator))
```
